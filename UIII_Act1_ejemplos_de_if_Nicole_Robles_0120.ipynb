#### Ejemplos de if,if else,if elif

#### Ejemplo de if

```python
# 1. Uso de if simple con operadores relacionales
# Este código verifica si un número es positivo.
numero = 15
if numero > 0:
    print("El número es positivo.")
```

```python
# 2. Uso de if else con operadores aritméticos y relacionales
# Este código determina si un número es par o impar utilizando el operador de módulo.
valor = 10
if valor % 2 == 0:
    print("El número es par.")
else:
    print("El número es impar.")
```

```python
# 3. Uso de if elif else con operadores lógicos y relacionales
# Este ejemplo clasifica a una persona según su edad utilizando el operador lógico and.
edad = 25
if edad < 13:
    print("Es un niño.")
elif edad >= 13 and edad < 20:
    print("Es un adolescente.")
else:
    print("Es un adulto.")
```

```python
# 4. Ejemplo complejo: Año bisiesto
# Este ejemplo combina operadores aritméticos, relacionales y lógicos (and, or) para una lógica más avanzada
anio = 2024
if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
    print(f"{anio} es un año bisiesto.")
else:
    print(f"{anio} no es bisiesto.")
```
